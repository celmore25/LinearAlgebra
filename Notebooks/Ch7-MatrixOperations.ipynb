# Matrix Operations

```python
import numpy as np
from scipy import linalg as la
from matplotlib import pyplot as plt
from scipy.signal import convolve2d
from scipy import misc
```

```python
def plot_2d_matrix(X, ax):
    assert X.shape == (2,2)
    ax.arrow(0, 0, X[:,0][0], X[:,0][1], head_width=0.05, length_includes_head=True, color='red', label='Original 1')
    ax.arrow(0, 0, X[:,1][0], X[:,1][1], head_width=0.05, length_includes_head=True, color='red', label='Original 2')
    ax.set_ylim([-1.1, 1.1])
    ax.set_xlim([-1.1, 1.1])
    ax.grid()
    x = np.linspace(-1,1,100)
    y = np.sqrt(1 - x**2)
    ax.plot(x, y, 'k')
    ax.plot(x, -y, 'k')
    return ax
```

## Geometric Transformation

```python
def rotate(X, theta):
    r = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
    return r@X
```

```python
X = np.random.uniform(-1, 1, (2,2))
fig, ax = plt.subplots(figsize = (5, 5))
ax = plot_2d_matrix(X, ax)
ax.set_title('Original')
plt.show()

fig, ax = plt.subplots(figsize = (5, 5))
ax.set_title('Rotated $\pi$ radians')
Xr = rotate(X, np.pi)
ax = plot_2d_matrix(Xr, ax)
plt.show()
```

## Convolution

### Vector

```python
x = np.linspace(-5, 5, 300)
y = np.sin(x) + np.random.uniform(-0.1, 0.1, 300)
c = np.convolve(y, np.array([1/10 for i in range(10)]))
plt.plot(x, y, label = 'Original Signal')
plt.plot(x, c[5:305], label = 'Smoothing Convolution')
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

## Matrix

```python
X = misc.face(gray=True)
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(X, cmap='gray')
ax.set_xticks([]); ax.set_yticks([])
ax.set_title('Original Image')
plt.show()

n = 20
c_map = np.array([[1/n for i in range(n)] for i in range(n)])
Xc = convolve2d(X, c_map)
fig, ax = plt.subplots(figsize = (8, 8))
ax.imshow(Xc, cmap='gray')
ax.set_xticks([]); ax.set_yticks([])
ax.set_title(f'Convolved Blurred Image n={n}')
plt.show()
```
